```python
#seu código aqui
#a)
def freq(texto):
    palavras = [x.lower().strip(string.punctuation) for x in texto.split()]
    dic = {}
    for item in palavras:
        if item in dic:
            dic[item] += 1
        else:
            dic[item] = 1
    return dic

freq(texto)
            
            

    
```

```python
#seu código aqui
#b)
def proprios(texto):
    palavras = [x.strip(string.punctuation) for x in texto.split()]
    Lista_maiusculas = [x for x in palavras if x.istitle() is True]
    
    #verifica se há pontuacao.
    L = []
    for x in Lista_maiusculas:
        posicao = texto.index(x)
        a = ' '
        while a == ' ':
            posicao -= 1
            a = texto[posicao-1]
        if a in '!?.':
            pass
        else:
            L.append(x)
    return set(L)        
    
    
```

```python
#seu código aqui
#c)Imagine que o texto seja constituído pela soma de suas letras, com repetições. Com esta "caixa de letras",
#quantas vezes você conseguiria escrever seu nome completo? Use as letras da variável meu_nome, sem os espaços em branco.

def nome():
    Lista_letras = [x for x in texto.lower()]
    Qtd = []
    nome = 'tarlalemosdeandrade'

#dicionario com as repeticoes de cada letra:
    dic={}
    for x in nome:
        if x in dic:
            dic[x] += 1
        else:
            dic[x] = 1

#lista com qunatas letras do meu nome ha no texto:    
    for x in dic.keys():
        a = Lista_letras.count(x)
        Qtd.append(a)
    
    
    Lista_values = [x for x in dic.values()]   
    qtd_letras = [round(x/y) for x,y in zip(Qtd,Lista_values)]

    K = min(qtd_letras)
    return K
    
    

```

```python
#seu código aqui
#d)Identifique as palavras que só ocorrem uma vez
def unico():
    unicas = []
    a = freq(texto)
    for x in a.keys():
        if a[x] == 1:
            unicas.append(x)
    return unicas
        
    
```

```python
#seu código aqui
#e)

with open('respostas_q1.txt','w') as t:
    freq = freq(texto)
    for key,value in freq.keys(),freq.values():
        t.write('{},{}\n'.format(key,value))
    t.write('\n')
    proprios = proprios(texto)
    for proprio in proprios:
        t.write('{}\n'.format(proprio))
    t.write('\n')
    
    for n in nome:
        t.write('{}\n'.format(n))
    t.write('\n')
    for u in unico:
        t.write('{}\n'.format(u))
        
    
```
