```python
# uniq -c (count) conta as ocorrências.
# head/tail
# tr (translate)
# tr -sc separa as palavras (split)
# ^(início)
# $(fim)
# * (tudo)
# sort |uniq  = sort -u
# sort -f (ignora case(maiúscula/minúscula))
# negar conjunto com grep: grep -v '[aeiou]'
#regex 101
```

```python
! cat ./Dados/genesis.txt | tr -sc '[A-Z][a-z]' '[\012*]' | sort | uniq -c | sort -gr | sed 10q
```

```python
# tr -s (squeeze-conjunto de caracteres seguidos-SET)
# tr -c (complemento - tudo o que não está no conjunto indicado)
# '[\012*]' (código para quebra de linha \n - new line)
# tr -sc '[A-Z][a-z]' '[\012*]' (pega o conjunto de caracteres diferente de '[A-Z][a-z]' e substitui por'[\012*]' que
  # representa newline - o '-s' é para dizer que é o conjunto todo- ex:' ...' - espaço e três pontos).

# sort -gr (-g indica ordenaação pelos valores numéricos e -r é reverse (decrescente))

#sed 10q (considerar apenas as 10 primeiras linhas).
```

```python
! cat ./Dados/genesis.txt | tr '[a-z]' '[A-Z]' | tr -sc '[A-Z]' '[\012*]' | sort | uniq | sed 10q
```

```python
! cat ./Dados/genesis.txt | tr '[a-z]' '[A-Z]' | tr -sc '[A-Z]' '[\012*]' |sort | uniq -c | sort -gr | sed 10q
```

```python
! ! cat ./Dados/genesis.txt | tr -sc '[A-Z][a-z]' '[\012*]' | rev |sort | rev | uniq -c | sed -n '100,120p'
```

#### 3) Encontrando n-gramas

```python
! cat ./Dados/genesis.txt | tr -sc '[A-Z][a-z]' '[\012*]' > ./Dados/genesis.words
! tail +2 ./Dados/genesis.words > ./Dados/genesis.nextwords
! paste ./Dados/genesis.words ./Dados/genesis.nextwords | sort | uniq -c > ./Dados/genesis.bigrams
```

```python
! cat ./Dados/genesis.bigrams | head 
```

```python
! paste ./Dados/genesis.words ./Dados/genesis.nextwords | sort | uniq -c | sort -gr > ./Dados/genesis.bigrams.freq
```

```python
! cat ./Dados/genesis.bigrams.freq | head 
```
